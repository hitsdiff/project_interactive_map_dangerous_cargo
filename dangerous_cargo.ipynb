### Тип данных

```python
df["Дата поступления"]=pd.to_datetime(df["Дата поступления"],format="%d.%m.%Y %H:%M")
```

```python
df["Наименование контакта"].fillna('Без названия', inplace=True)
df["Наименование контакта"]= df["Наименование контакта"].astype('string')

print("Количество клиентов до преобразования: ",len(df["Наименование контакта"].unique()))
df["Наименование контакта"]= df["Наименование контакта"].apply(lambda x: x.upper().replace("  "," ").replace('«','').replace('»','').replace('"','').replace("'",""))
print("Количество клиентов после преобразования",len(df["Наименование контакта"].unique()))
```

```python
def check_data(x: str):
  x=str(x)
  match=re.search(r'\W.\d\d.\d\d.\d\d\d\d.\d\d.\d\d.\W.\w\w\w.\W',x)
  date=match[0] if match else 'Not found'
  if date!="Not found":
    date=date.split("-")
    return(date[1])
```

```python
df["Дата закрытия"]=df["Решение, направленное клиенту"].apply(lambda x: check_data(x))
```

```python
df['Суть обращения']=[x.upper() for x in df['Суть обращения']]
df['Суть обращения']=[string.replace('\n'," ") for string in df['Суть обращения']]
df['Суть обращения']=[string.replace('\xa0'," ") for string in df['Суть обращения']]
```

```python
out=['МАГАЗИНОВ','ПОРОХИН','КЕРОСИНСКИЙ','КУРАНАХ']
danger_list=[]
df_danger=pd.DataFrame()
counter=0
for x in df['Суть обращения']:
    result=re.sub(r'[^\w]', ' ', x)
    result=re.sub(r'[a-zA-Z\s]', ' ', result)
    #print(result)
    result =re.findall(pattern, result)

    #для станции Бензин
    index_benzin=x.find('БЕНЗИН')
    if (x[index_benzin-8:index_benzin+6]=='СТАНЦИЮ БЕНЗИН' or x[index_benzin-8:index_benzin+6]=='СТАНЦИИ БЕНЗИН' or x[index_benzin-8:index_benzin+6]=='СТАНЦИЯ БЕНЗИН' or x[index_benzin+9:index_benzin+12]=='КБШ' or x[index_benzin-5:index_benzin-1]=='ВЧДР' or x[index_benzin-4:index_benzin-1]=='ВРД') and len(unique(result))==1 :
        result=[]

    #для станции Азотная
    index_azot=x.find('АЗОТНАЯ')
    if ((x[index_azot-8:index_azot+7]=='СТАНЦИЮ АЗОТНАЯ' or x[index_azot-8:index_azot+7]=='СТАНЦИИ АЗОТНАЯ' or x[index_azot-8:index_azot+7]=='СТАНЦИЯ АЗОТНАЯ' or x[index_azot+8:index_azot+11]=='МСК')  and len(unique(result))==1 ) or (x[index_azot-4].isdigit()==True):
        result=[]

    #для организаций
    for i in result:
        index_org=x.find(i)
        if (x[index_org-4:index_org-2]=='АО' or x[index_org-5:index_org-2]=='ОАО' or x[index_org-5:index_org-2]=='ООО') and (len(unique(result))==1):
            result=[]

    for i in out:
        for j in result:
            if j==i and len(unique(result))==1:
                result=[]

    if (len(result)!=0):
        danger_list.append(x)
        df_danger = df_danger.append(df.iloc[counter], ignore_index=True)
        # index=x.find(result[0])
        # print(x[index-40:index+40])
        #print(result)
        # print(counter)
        # print('----------------------------------------------------------------------------------------')
    # counter=counter+1
print(len(danger_list))
```
